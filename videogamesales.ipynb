```python
import numpy as np
import pandas as pd
import collections
import matplotlib.pyplot as plt

data = pd.read_csv("vgsales.csv", index_col='Rank')
```

```python
data
```

```python
eu = []
na = []
jp = []
for index, row in data.iterrows():
    bigest = max(row['NA_Sales'], row['EU_Sales'], row['JP_Sales'])
    if bigest == row['NA_Sales']:
        na.append(row['Genre'])
    elif bigest == row['EU_Sales']:
        eu.append(row['Genre'])
    elif bigest == row['JP_Sales']:
        jp.append(row['Genre'])
dict1 = collections.Counter(eu)
plt.pie(dict1.values(),labels=dict1.keys(), radius = 2, autopct='%0.0f%%', startangle=90)
plt.title("EU most sold genres", pad=150)
plt.show()
```

```python
dict2 = collections.Counter(na)
plt.pie(dict2.values(), labels=dict2.keys(), radius = 2, autopct='%0.0f%%', startangle=90)
plt.title("NA most sold genres", pad=150)
plt.show()
```
